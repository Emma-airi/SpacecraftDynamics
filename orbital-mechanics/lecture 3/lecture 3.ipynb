## The Eccentricity Vector

### Deriving The Eccentricity Vector

The Eccentricity vector $\vec{e}$ helps us describe the shape of an orbit and its direction. We can find it by taking the cross product of the velocity and the angular momentum

1. Such as:

   $$
   \ddot{\vec{r}} = - \frac{\mu}{r^3} \vec{r}
   $$

   Taking the cross product with the specific angular momentum, $\vec{h}$:

   $$
   \ddot{\vec{r}}\times\vec{h} = \textcolor{red}{- \frac{\mu}{r^3} \vec{r}}\times\vec{h}
   $$
   Using the rule for $\vec{a}\times\vec{b}\times{c}=(\vec{c}\cdot\vec{a})\vec{b}-(\vec{b}\cdot\vec{a})\vec{c}$ with $\vec{h}=\vec{r}\cdot\dot{\vec{r}}$:
$$
-\frac{\mu}{r^3}[(\dot{\vec{r}}\cdot\vec{r})\vec{r}-(\vec{r}\cdot\vec{r})\dot{\vec{r}}]\rightarrow-\frac{\mu}{r^3}(\dot{r}r\vec{r}-r^2\dot{\vec{r}})\rightarrow-\frac{\mu}{r^2}(\dot{r}\vec{r}-r\dot{\vec{r}})\\\Rightarrow =\frac{d}{dt}(\mu\frac{\vec{r}}{r})
$$
   This gives us:

   $$
   \Rightarrow \dot{\vec{r}}\times\vec{h} - \mu\frac{\vec{r}}{r} = constant
   $$
   Which can be rewritten as:
   $$
   \bold{\vec{e}=\frac{\dot{\vec{r}}\times\vec{h}}{\mu}-\frac{\vec{r}}{r}}
   $$
   This is our equation for the eccentricity vector!

2. The first integrals of motion:

   - $\vec{e}$ is a constant vector, implying both constant direction and magnitude.
   - $\vec{e}$ lies in the plane of motion and is fixed.

   Let's define:

   $$
   \bold{e} = \frac{\vec{e}}{\mu}
   $$

   where $e$ is the eccentricity vector.
### Properties

* $\vec{e}$ lies in the plane of motion and (since it is constant) has a fixed magnitude and direction.
   $\Rightarrow \hat{i}\equiv\vec{e}$, so we can use the eccentricity as one of the axes.

### b. The Equation of the Orbit

1. Recall the expression:

   $$
   \mathbf{r} = \frac{p}{1 + e \cos(\theta)}
   $$

   where $\theta$ is the true anomaly.

2. Different conic sections are described depending on the value of $e$:

   - $e = 0$: Circular orbit
   - $0 < e < 1$: Elliptic orbit
   - $e = 1$: Parabolic orbit
   - $e > 1$: Hyperbolic orbit

   Operative orbits are always circular or elliptic.

### a. Key Parameters

1. Semi-latus rectum $p$ and semi-major axis $a$:

   $$
   p = a(1 - e^2)
   $$

2. Perigee ($r_p$):

   $$
   r_p = a(1 - e)
   $$

3. Apogee ($r_a$):

   $$
   r_a = a(1 + e)
   $$

## Vis Viva Equation

1. The Vis Viva equation:

   $$
   v^2 = \mu \left(\frac{2}{r} - \frac{1}{a}\right)
   $$

   - $v$ is the orbital speed.
   - $r$ is the distance from the focal point to the orbiting body.
   - $a$ is the semi-major axis.

2. This equation is extremely useful in mission design for calculating the velocity at a given point in the orbit when planning maneuvers.

## Specific Orbital Energy

1. The specific orbital energy E:

   $$
   E = \frac{v^2}{2} - \frac{\mu}{r}
   $$

   - e is a constant for any given orbit.
   - For elliptical orbits, $E < 0$.
   - For parabolic orbits, $E = 0$.
   - For hyperbolic orbits, $E > 0$.

2. Orbits with the same semi-major axis $a$ have the same energy.

## Conclusion

- The eccentricity vector is a fundamental constant vector in orbital mechanics.
- The orbit equation describes different conic sections based on the value of the eccentricity $e$.
- The Vis Viva equation is a crucial tool in mission planning for calculating velocities at various points in an orbit.
- The specific orbital energy determines the nature of the orbit, whether it is circular, elliptical, parabolic, or hyperbolic.
