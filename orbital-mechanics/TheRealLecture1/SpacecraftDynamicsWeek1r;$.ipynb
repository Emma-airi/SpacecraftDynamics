### Lecture 1

# Vectors (Brief Recap)

- A vector is a quantity with a direction and a magnitude.
- $\vec{r}$, $\mathbf{r}$, $\boldsymbol{r}$ are all possible notations for vector quantities.
- We describe a vector by its components:
  \
  $\vec{r}$ = $r_x \hat{\imath}$ + $r_y \hat{\jmath}$ + $r_z \hat{k}$ = ($r_x,  r_y,  r_z$)
  
  where $\hat{\imath}$, $\hat{\jmath}$, and $\hat{k}$ are unit vectors aligned with a triad of orthogonal axes.

- The magnitude is indicated by $\|\vec{r}\|$ or simply by $r$ and
  \
  r = $\sqrt{r_x^2 + r_y^2 + r_z^2}$
  


# Vector Operations

## Multiplication of a Scalar with a Vector

- When a vector $\vec{A}$ is multiplied by a scalar $Q$, the result is a new vector $Q\vec{A}$, which has the same direction as $\vec{A}$ but a magnitude that is $|Q|$ times the magnitude of $\vec{A}$.

  \
  $Q \vec{A}$ = $Q (A_x \hat{\imath}$ + $A_y \hat{\jmath}$ + $A_z \hat{k}$) = $(Q A_x) \hat{\imath}$ + $(Q A_y) \hat{\jmath}$ + $(Q A_z) \hat{k}$
  

## Sum of Vectors

- The sum of two vectors $\vec{A}$ and $\vec{B}$ is a vector $\vec{C}$ such that:
  \
  $\vec{C}$ = $\vec{A}$ + $\vec{B}$ = $(A_x + B_x) \hat{\imath}$ + $(A_y + B_y) \hat{\jmath}$ + $(A_z + B_z) \hat{k}$
  

## Parallelogram Rule

- The sum of two vectors can also be represented geometrically using the parallelogram rule:
  - Place the vectors $\vec{A}$ and $\vec{B}$ tail to tail.
  - Draw a parallelogram where $\vec{A}$ and $\vec{B}$ are adjacent sides.
  - The interception of the two projected lines represents the point where A is summed with B, $\vec{A} + \vec{B}$.



## Dot Product Rule

- The dot product (or scalar product) of two vectors $\vec{A}$ and $\vec{B}$ is defined as:
  \
  $\vec{A} \cdot \vec{B}$ = $A_x B_x + A_y B_y + A_z B_z$ = $|\vec{A}| |\vec{B}| \cos(\theta)$
  \
  where $\theta$ is the angle between $\vec{A}$ and $\vec{B}$.

- If one of the vectors is a unit vector (say, $\hat{B}$), the dot product gives the projection of $\vec{A}$ onto $\hat{B}$:
  \
 $ \vec{A} \cdot \hat{B}$ = $|\vec{A}| \cos(\theta)$
  \
  This is because $|\hat{B}| = 1$.



# Vector Operations: Cross Product and Related Formulas

## Cross Product

- The cross product of two vectors $\vec{A}$ and $\vec{B}$, denoted $\vec{A} \times \vec{B}$, results in a vector $\vec{C}$ that is perpendicular to both $\vec{A}$ and $\vec{B}$.
- The magnitude of $\vec{A} \times \vec{B}$ is given by:
  \[
  |$\vec{A} \times \vec{B}|$ = $|\vec{A}| |\vec{B}| \sin(\theta)$
  \]
  where $\theta$ is the angle between $\vec{A}$ and $\vec{B}$.
- The direction of $\vec{A} \times \vec{B}$ is determined by the right-hand rule.

\
$\vec{A} \times \vec{B}$ = \begin{vmatrix}
\hat{\imath} & \hat{\jmath} & \hat{k} \\
A_x & A_y & A_z \\
B_x & B_y & B_z
\end{vmatrix}
= $\left( A_y B_z - A_z B_y \right)$ $\hat{\imath}$ - $\left( A_x B_z - A_z B_x \right)$ $\hat{\jmath}$ + $\left( A_x B_y - A_y B_x \right)$ $\hat{k}$


### Properties of the Cross Product

1. **Anticommutative Property:**
   \[
   $\vec{A} \times \vec{B}$ = $- (\vec{B} \times \vec{A})$
   \]

2. **Distributive Property:**
   \[
   $\vec{A} \times (\vec{B} + \vec{C})$ = $\vec{A} \times \vec{B} + \vec{A} \times \vec{C}$
   \]

3. **Scalar Multiplication:**
   \[
   $(k \vec{A}) \times \vec{B} = k (\vec{A} \times \vec{B}) = \vec{A} \times (k \vec{B})$
   \]

### Triple Vector Product

- The triple vector product involves three vectors $\vec{A}$, $\vec{B}$, and $\vec{C}$:

  \
  $\vec{A} \times (\vec{B} \times \vec{C})$ = $(\vec{A} \cdot \vec{C}) \vec{B}$ - $(\vec{A} \cdot \vec{B}) \vec{C}$
  \
  or\
  \
  $(\vec{A} \times \vec{B}) \times \vec{C}$ = $(\vec{C} \cdot \vec{B}) \vec{A}$ - $(\vec{C} \cdot \vec{A}) \vec{B}$
  

### Scalar Triple Product

- The scalar triple product of three vectors $\vec{A}$, $\vec{B}$, and $\vec{C}$ is a scalar quantity:
  \
  $\vec{A} \cdot (\vec{B} \times \vec{C}$)
  
- This scalar represents the volume of the parallelepiped formed by the vectors $\vec{A}$, $\vec{B}$, and $\vec{C}$.

  \
  $\vec{A} \cdot (\vec{B} \times \vec{C})$ = \begin{vmatrix}
  A_x & A_y & A_z \\
  B_x & B_y & B_z \\
  C_x & C_y & C_z
  \end{vmatrix}

  \
  $\vec{A} \cdot (\vec{B} \times \vec{C})$ = $\vec{B} \cdot (\vec{C} \times \vec{A})$ = $\vec{C} \cdot (\vec{A} \times \vec{B})$
  

# Time Derivative of a Vector

- The time derivative of a vector $\vec{A}$ represents how the vector changes with respect to time.

### In an Inertial Reference Frame


- The time derivative of a vector $\vec{r}$ in an inertial reference frame is given by:

\
$\frac{d\vec{r}}{dt}$ = $\frac{d}{dt}(x) \mathbf{i}$ + $\frac{d}{dt}(y) \mathbf{j}$ + $\frac{d}{dt}(z) \mathbf{k}$



### In a Rotating Reference Frame

- In a rotating reference frame with angular velocity $\boldsymbol{\omega}$, the time derivative of the vector $\vec{r}$ must account for the rotation:
  \
  $\left( \frac{d\vec{r}}{dt} \right)_{\text{inertial}}$ = $\left( \frac{d\vec{r}}{dt} \right)_{\text{rotating}}$ + $\boldsymbol{\omega} \times \vec{r}$
  
  where:
  - $\left( \frac{d\vec{r}}{dt} \right)_{\text{rotating}}$ is the time derivative in the rotating frame
  - $\left( \frac{d\vec{r}}{dt} \right)_{\text{inertial}}$ is the time derivative in the inertial frame
  - $\boldsymbol{\omega}$ is the angular velocity of the rotating frame
  - $\times$ denotes the cross product (see above for more information)
  - This can also be expressed as:
  \
  $\left( \frac{d\vec{r}}{dt} \right)_{\text{rot}}$ = $\left( \frac{d\vec{r}}{dt} \right)_{\text{inertial}}$  - $\boldsymbol{\omega} \times \vec{r}$
  
  




# Gravitational Force and Newton's Second Law

- The gravitational force between two masses $\ m_1 \ and \ m_2 \$ is given by:

### Gravitational Force
\
\begin{aligned}
\vec{F}_{12} &= \frac{G m_1 m_2}{r^2} \hat{r} = \frac{G m_1 m_2}{r^3} \vec{r} \\
\vec{F}_{21} &= -\frac{G m_1 m_2}{r^2} \hat{r} = -\frac{G m_1 m_2}{r^3} \vec{r}
\end{aligned}


where:
- \$ G = 6.672 \times 10^{-11} \ \text{kg}^{-1} \ \text{m}^3 \ \text{s}^{-2} \$ (Universal gravitational constant)
- $\\hat{r} = \frac{\vec{r}}{|\vec{r}|} \$ is the unit vector



- $\vec{F}_{12}$ Represents the force on $\ m_1 \ exerted  by  \ m_2 \$'s gravitational attraction.

- $\vec{F}_{21}$ Represents the force on $\ m_2 \ exerted  by  \ m_1 \$'s gravitational attraction.


### Center of Mass \( G \)
The center of mass $\ \vec{r}_G \$ of the system is given by:


$\vec{r}_G$ = $\frac{m_1 \vec{r}_1 + m_2 \vec{r}_2}{m_1 + m_2}
\$


### Newton's Second Law (F = ma)
1. $ m_1 \ddot{\vec{r}}_1 = \vec{F}_{12} $
2. $ m_2 \ddot{\vec{r}}_2 = \vec{F}_{21} $

Combining these equations:

$\
m_1 \ddot{\vec{r}}_1$ + $m_2 \ddot{\vec{r}}_2$ = $\vec{F}_{12}$ + $\vec{F}_{21}$ = $\frac{G m_1 m_2 \vec{r}}{r^3} $ + $-\frac{G m_1 m_2 \vec{r}}{r^3} $ = 0 (Reference Equations Above)

Thus:

$\
\ddot{\vec{r}}_G = 0, \quad \dot{\vec{r}}_G = \text{const}
\$

This implies that $ G $ can be chosen as the origin of the inertial frame.


### Two-Body Relative Dynamics

#### Newton's Second Law
1. $ m_2 \ddot{\vec{r}}_2 = \vec{F}_{21} = -  \frac{G m_1 m_2 \vec{r}}{r^3}  \rightarrow  \ddot{\vec{r}}_2 = -  \frac{G m_1  \vec{r}}{r^3} $

2. $ m_1 \ddot{\vec{r}}_1 = \vec{F}_{12} =  \frac{G m_1 m_2 \vec{r}}{r^3} \rightarrow  \ddot{\vec{r}}_2 =  \frac{G m_2  \vec{r}}{r^3} $

Combining these equations:

$
\ddot{\vec{r}} = \ddot{\vec{r}}_2 - \ddot{\vec{r}}_1 = - G \frac{m_1 \vec{r}}{r^3} - G \frac{m_2 \vec{r}}{r^3} = - G \frac{(m_1 + m_2) \vec{r}}{r^3}
$

$
\ddot{\vec{r}} = - \frac{G (m_1 + m_2)}{r^3} \vec{r}
$

$
\ddot{\vec{r}} = - \frac{\mu \vec{r}}{r^3}
$

where $\\mu\$ is the gravitational parameter.

#### Implications of the Equation of Motion

- If $\ m_1 \gg m_2 $

  e.g., $ m_1 $ is $ m_\oplus = 5.974  \times 10^{24}  kg$  $\ and\   m_2\ is\   a\   spacecraft\   m_2 \approx 1000  kg$

$
\vec{r}_G \approx \vec{r}_1
$

$
\mu \approx G m_1
$

This is referred to as a restricted two-body problem.
